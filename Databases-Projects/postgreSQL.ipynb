```python
#The query below is an outer join query that combines the pagila.customer and pagila.payment tables, where customer.customer_id and payment.customer_id are equal, given that the payment amount is greater than or equal to 5. It then orders it by customer first name. 
```

```python
!psql -c "SELECT * FROM pagila.customer FULL OUTER JOIN pagila.payment ON customer.customer_id=payment.customer_id WHERE payment.amount >=5  ORDER BY customer.first_name LIMIT 10"
```

```python
#The query below is an outer join query that combines the pagila.customer and pagila.payment tables, where customer.customer_id and payment.customer_id are equal, given that the payment amount is greater than or equal to 10. It then orders it by customer first name.
```

```python
#The query below is an outer join query that combines the pagila.customer and pagila.payment tables, where customer.customer_id and payment.customer_id are equal, given that the customer.customer_id is greater than 300 & 400. It then orders it by customer first name.
```

```python
!psql -c "SELECT * FROM pagila.customer FULL OUTER JOIN pagila.payment ON customer.customer_id=payment.customer_id WHERE 300 < customer.customer_id AND customer.customer_id > 400 ORDER BY customer.first_name LIMIT 10"
```
