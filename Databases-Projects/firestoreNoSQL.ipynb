```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
batch = db.batch()

df = pd.read_csv('shopify/apps.csv', sep=',', header=0, lineterminator='\n')
apps_rows = df.values.tolist()

df = pd.read_csv('shopify/apps_categories.csv', sep=',', header=0, lineterminator='\n')
apps_categories_rows = df.values.tolist()

df = pd.read_csv('shopify/categories.csv', sep=',', header=0, lineterminator='\n')
categories_rows = df.values.tolist()



for apps_row in apps_rows:
    apps_record = {}
    apps_record['id'] = apps_row[0]
    apps_record['url'] = apps_row[1]
    apps_record['title'] = apps_row[2]
    apps_record['developer'] = apps_row[3]
    apps_record['developer_link'] = apps_row[4]
    apps_record['icon'] = apps_row[5]
    apps_record['rating'] = apps_row[6]
    apps_record['reviews_count'] = apps_row[7]
    
    apps_ref = db.collection('apps').document(apps_row[0])
    
    batch.set(apps_ref, apps_record)
    
    for apps_categories_row in apps_categories_rows:
        if apps_row[0] == apps_categories_row[0]:
            apps_categories_record = {}
            apps_categories_record['category_id'] = apps_categories_rows[1]
            
            apps_categories_ref = apps_ref.collection('apps_categories').document(apps_categories_row[1])
        
            batch.set(apps_categories_ref, apps_categories_record)
            batch.commit()
            
    for categories_row in categories_rows:
        if apps_categories_row[1] == categories_row[0]:
            categories_record = {}
            categories_record['title'] = categories_rows[1]
            
            categories_ref = apps_categories_ref.collection("categories").document(categories_row[1])
            batch.set(categories_ref, categories_record)
            batch.commit()
            
            
    
```

```python
db = firestore.Client()
batch = db.batch()

df = pd.read_csv('shopify/apps.csv', sep=',', header=0, lineterminator='\n')
apps_rows = df.values.tolist()

df = pd.read_csv('shopify/pricing_plans.csv', sep=',', header=0, lineterminator='\n')
pricing_plans_rows = df.values.tolist()

df = pd.read_csv('shopify/pricing_plan_features.csv', sep=',', header=0, lineterminator='\n')
pricing_plan_features_rows = df.values.tolist()

for row in apps_rows: 
    record = {}
    record['id'] = row[0]
    record['url'] = row[1]
    record['title'] = row[2]
    record['developer'] = row[3]
    record['developer_link'] = row[4]
    record['icon'] = row[5]
    record['rating'] = row[6]
    record['reviews_count'] = row[7]
    
    apps_ref = db.collection('apps').document(row[0])
    batch.set(apps_ref, record)
    
    batch.commit()
    
    for pricing_plans_row in pricing_plans_rows:
        if row[0] == pricing_plans_row[1]:
            pricing_plans_record = {}
            pricing_plans_record['title'] = pricing_plans_row[2]
            pricing_plans_record['price'] = pricing_plans_row[3]
            
            pricing_plans_ref = apps_ref.collection('pricing_plans').document(pricing_plans_row[1])
            batch.set(pricing_plans_ref, pricing_plans_record)
            batch.commit()
            
    for pricing_plan_features_row in pricing_plan_features_rows:
        if row[0] == pricing_plan_features_row[0]:
            pricing_plan_features_record = {}
            pricing_plan_features_record["pricing_plan_id"] = pricing_plan_features_row[1]
            pricing_plan_features_record["feature"] = pricing_plan_features_row[2]
            
            pricing_plan_features_ref = apps_ref.collection('pricing_plan_features').document(pricing_plan_features_row[0])
            batch.set(pricing_plan_features_ref, pricing_plan_features_record)
            batch.commit()
    
    
```

```python
db = firestore.Client()
batch = db.batch()

df = pd.read_csv('shopify/apps.csv', sep=',', header=0, lineterminator='\n')
apps_rows = df.values.tolist()

df = pd.read_csv('shopify/key_benefits.csv', sep=',', header=0, lineterminator='\n')
key_benefits_rows = df.values.tolist()

for row in apps_rows: 
    record = {}
    record['id'] = row[0]
    record['url'] = row[1]
    record['title'] = row[2]
    record['developer'] = row[3]
    record['developer_link'] = row[4]
    record['icon'] = row[5]
    record['rating'] = row[6]
    record['reviews_count'] = row[7]
    
    apps_ref = db.collection('apps').document(row[0])
    batch.set(apps_ref, record)
    
    batch.commit()
    
    for key_benefits_row in key_benefits_rows:
        if row[0] == key_benefits_row[0]:
            key_benefits_record = {}
            key_benefits_record['title'] = key_benefits_row[1]
            key_benefits_record['description'] = key_benefits_row[2]
            
            key_benefits_ref = apps_ref.collection("key_benefits").document(key_benefits_row[0])
            batch.set(key_benefits_ref, key_benefits_record)
            batch.commit()
```
